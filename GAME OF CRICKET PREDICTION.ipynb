#### *Train/Test Split involves splitting the dataset into training and testing sets respectively, which are mutually exclusive. After which, you train with the training set and test with the testing set.
#### This will provide a more accurate evaluation on out-of-sample accuracy because the testing dataset is not part of the dataset that have been used to train the data. It is more realistic for real world problems.*

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

* ### Train data distribution

```python
plt.scatter(train.OppositionTeam, train.Runs,  color='blue')
plt.xlabel("Opposition")
plt.ylabel("Runs")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['OppositionTeam']])
y = np.asanyarray(train[['Runs']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercepts: ', regr.intercept_)
```

 ### Coefficient and Intercept , are the parameters of the fit line.

```python
y_hat= regr.predict(test[['OppositionTeam']])
x = np.asanyarray(test[['OppositionTeam']])
y = np.asanyarray(test[['Runs']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
